```python
#Dropping the class feature
diabetes_df.drop("Class",1,inplace=True)
```

```python
#import the libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#pisahkan antara atribut dan label
X = diabetes_df.drop('Type',1)
y = diabetes_df.iloc[:,-1]
```

```python
from sklearn.model_selection import KFold
kf_3 = KFold(n_splits=3)
kf_5 = KFold(n_splits=5)
kf_10 = KFold(n_splits=10)
```

```python
for train_index, test_index in kf_3.split([1,2,3,4,5,6,7,8,9]):
    print(train_index, test_index)
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
scores_rf = []
for train_index, test_index in kf_5.split(X):
    #print(train_index, test_index)
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=100), X_train, X_test, y_train, y_test))
    
```

```python
scores_rf
```

```python
#Check the prediction precision and accuracy
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report # evaluasi performa model

for train_index, test_index in kf_5.split(X):
    #print(train_index, test_index)
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    model_rf = RandomForestClassifier(n_estimators=100)
    model_rf.fit(X_train,y_train) # training model/classifier
    y_pred = model_rf.predict(X_test) # melakukan prediksi
    
    print("confusion matrix: \n", confusion_matrix(y_test, y_pred))
    print(classification_report(y_test,y_pred))
    
    print("=======================================================")
```
